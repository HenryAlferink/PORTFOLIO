In this notebook we do the data-worth stuff for the source reconstruction problem.

```python
import numpy as np
from scipy.linalg import sqrtm
import matplotlib.pyplot as plt
import matplotlib
from corner import corner
import IPython
import helper as h

# plt.rcParams.update({'font.size': 25})
matplotlib.rcdefaults()

# for outputting images to .pgf format
matplotlib.rcParams.update({
    'font.family': 'serif',
    'text.usetex': True,
    'pgf.rcfonts': False,
})

%load_ext autoreload
%autoreload 2
```

#### Parameters (change these!)

```python
# random seed
np.random.seed(10)

# data standard deviation
data_std = 0.1

# prior standard deviation
prior_std = 1

# number of parameters:
n = 100 

# number of initial data points (there will be some additional points added later)
num_initial_points = 3

T = 300 # time at which we want to evaluate the concentration as a function of distance

infeasible = 65 # value below which cost will be infinite

# choose which optimality criterion to run with ("A" or "D")
optimality = "A"

# times to run entire algorithm 
iterations = 18 # number of iterations, including the "no additional data" iteration

# cost function to use ("zero" or "step" or "lin" or "pol")
cost_function_to_use = "lin"

# criteria for choosing x_optimal (either "phi_and_P" or "phi", the first is for uncertainty+cost, and the second
# is for uncertainty only.)
opt_criteria = "phi_and_P"
```

#### Creating true parameter set $\mathbf{m}_{true}$

```python
# independent variable corresponding to parameter set:
t = np.linspace(0,250,n)

# distance between time steps:
dt = t[-1] - t[-2] 
```

```python
# Ollie's version...
m_true = 3 + np.sin(2*np.pi*t/n)*np.sin(0.01*t)
```

#### Creating prior parameter set $\mathbf{m}_{prior}$

```python
# constant prior:
m_prior = np.ones(t.shape) * 3
```

```python
# comparing our prior model and the "true" model
plt.figure(figsize=(6,4))
plt.plot(t, m_true, "--", label="True model");
plt.plot(t, m_prior, label="Prior model");
plt.xlabel("Time (t)");
plt.ylabel("Source Concentration ($C_{in}$)");
plt.title("Prior model and true model")
plt.legend();

plt.close()
```

#### Prior covariance matrix

```python
C_prior = prior_std**2 * np.eye(n)
std = np.ones(n) * np.sqrt(prior_std)
Std = np.outer(std, std)
char_L = 10 # characteristic length
def exp_squared(i,j,L): 
    return np.exp(-(i-j)**2/(2*L**2))
C_prior = np.fromfunction(exp_squared, (n,n), L=char_L) * Std + np.eye(n)*1e-4

# plt.imshow(C_prior);
```

#### Forward model

```python
# forward operator parameters
D = 1; v = 1

def kernel(x, t): 
    return x / (2*np.sqrt(np.pi*D*t**3)) * np.exp(-(x-v*t)**2 / (4*D*t))

# construct design matrix. Note that this matrix is a function of T, the time we observe the data vector.
def design(x, t):
    m = x.size; n = len(t)
    G = np.zeros((m, n))
    for i in range(m):
        for j in range(n):
            G[i,j] = kernel(x[i], T-t[j]) * dt
    return G
```

#### Set of measurement locations $\mathcal{X}$

```python
# initialize data class, starting with an assumed linear spacing of already-known data points 
data = h.data(
#     x_lo=20, 
#     x_hi=300,
    x_lo=infeasible+15, 
    x_hi=290,
    num_data_points=num_initial_points, 
    design=design, 
    t=t, 
    m_true=m_true, 
    m_prior=m_prior,
    data_std=data_std,
    C_prior=C_prior,
    additional_x=None
)
```

```python
# New possible measurement locations to add:
newpoints = np.linspace(0,300,100)

# add new possible measurement locations one by one. This also automatically computes the 
# corresponding concentration data at that location
for i in range(len(newpoints)):
    data.add_points([newpoints[i]])
```

#### Measurement location cost function

```python
# make sure to give this function the same arguments given here

if cost_function_to_use=="step":
    def cost_function(x, infeasible):
        """ Function that returns a vector of cost values associated with the given input measurement locations."""
        c = np.zeros(x.shape)
        number_infeasible = (x<=infeasible).sum() # number of values in vector below infeasible point
        c[x<infeasible] = 99999 # infinite cost for x-values below a certain distance
        remaining = len(x) - number_infeasible # remaining number of locations in vector   

        curr_pos = number_infeasible
        levels = np.array([3,2,1])*0.55 # different possible cost values
        shift = np.floor(remaining / len(levels)).astype('int')
        for i in range(len(levels)):
            c[curr_pos:] = levels[i]
            curr_pos += shift
        return c
    
if cost_function_to_use=="lin":
    # zero cost function
        def cost_function(x, infeasible):
            if optimality=="D":
                c = -0.01*np.array(x) + 4.2
            if optimality=="A":
                c = -0.0005*np.array(x) + 0.2
            c[c<0] = 0
#             c[x<infeasible] = 20
            c[x<infeasible] = np.NaN
            return c  
    
if cost_function_to_use=="pol":
    # zero cost function
        def cost_function(x, infeasible):
            if optimality=="D":
                c = 4.942308 - 0.04*np.array(x) + 0.00009615385*np.array(x)**2
            if optimality=="A":
                c = -0.0005*np.array(x) + 0.1
            c[c<0] = 0
#             c[x<infeasible] = 20
            c[x<infeasible] = np.NaN
            return c  

if cost_function_to_use=="zero":
    # zero cost function
        def cost_function(x, infeasible):
            return np.ones(x.shape)*0
```

#### Evalutate cost of current system of measurments

```python
costs = []
uncertainty = [] # store uncertainty objective value for each iteration

# note that the first cost in this vector corresponds to the "no additional data case"
costs.append(np.sum(cost_function(data.X, infeasible)))
if optimality=="A":
    uncertainty.append(data.A_optimality_initial)
if optimality=="D":
    uncertainty.append(data.D_optimality_initial)    
```

```python
# initialize additional points list
x_optimal = []

start_idx = (newpoints<infeasible+15).sum()
skip = len(newpoints[newpoints<infeasible+15]) # this is to couteract the indices that have been skipped

# Choose x_optimal
if optimality=="A":
    if opt_criteria=="phi_and_P":
        idx = np.argmin(data.A_optimality[start_idx:])    
    elif opt_criteria=="phi":
        idx = np.argmin(data.A_optimality_nocost[start_idx:])
    idx += skip # this is to couteract the indices that have been skipped
    if type(idx)==np.int64:
        x_optimal.append(newpoints[idx])
    else:
        x_optimal.append(newpoints[idx[0]])
        
if optimality=="D":
    if opt_criteria=="phi_and_P":
        idx = np.argmin(data.D_optimality[start_idx:])    
    elif opt_criteria=="phi":
        idx = np.argmin(data.D_optimality_nocost[start_idx:])  
    idx += skip # this is to couteract the indices that have been skipped
    if type(idx)==np.int64:
        x_optimal.append(newpoints[idx])
    else:
        x_optimal.append(newpoints[idx[0]])
```
