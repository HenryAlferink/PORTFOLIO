```python
import numpy as np
from scipy.linalg import sqrtm
import matplotlib.pyplot as plt
import matplotlib
from corner import corner
import IPython

np.random.seed(10)

# plt.rcParams.update({'font.size': 18})
matplotlib.rcdefaults()
```

## Ballistics Example
### Setup
We want to solve $\mathbf{d} = \mathbf{Gm}$ for $\mathbf{m}$ with the regression model being
$$y(t) = m_1 + m_2 t - \frac{1}{2}m_3 t^2$$

For example purposes, we will use $\mathbf{m}_{true}$ to be

```python
m_true = np.array([10, 100, 9.8])
```

And we'll set the time steps to be

```python
m = 25
```

```python
t = np.arange(1,m+1)
print('e.g.')
t[:10]
```

With $m_{true}$ we can create some synthetic data for different time steps. We add a bit of noise to each data point; i.e. we are using
$$d_i = y(t_i) + \epsilon_i$$
where $\epsilon_i \overset{iid}{\sim} N(0,64)$. Note that we make the synthetic data from samples taken from a normal distribution around the *true* model.

```python
data_stddev = 70 # standard deviation of data points
# note that a data_stddev=8 is what the example in Aster uses.
```

```python
def f(time, model): return model[0] + model[1]*time - 0.5*model[2]*time**2 

d = f(t, m_true) + np.random.normal(0, data_stddev, m) # the second argument is std. deviation, not variance.
```

```python
plt.figure(figsize=(7,5))
plt.plot(t, f(t, m_true), label="True Model");
plt.errorbar(x=t, y=d, yerr=data_stddev, fmt=".", capsize=3, label="Synthetic Data (with 1$\sigma$ error bars)");
plt.xlabel('Time');
plt.ylabel('y');
plt.legend(loc="lower left");
```

We now construct the $G$ matrix as follows
$$G_{i,.} = [1, t_i, -(1/2)t_i^2]$$

```python
G = np.zeros((m, 3))
for i in range(m):
    G[i,:] = [1, t[i], -0.5*t[i]**2]
print('e.g.')
G[:5, :]
```

### Solving using the Bayesian framework
We now move on to finding the posterior. We assume that the data is iid normal distributed. So,
$$C_D = \sigma_d^2 \mathbf{I}_m.$$

```python
C_D = data_stddev**2 * np.eye(m)
print('e.g.')
C_D[:5,:5]
```

Furthermore, we will take the prior to be 

```python
m_prior = np.array([5, 110, 5])
```

and we set the prior covariance matrix such that variables are independent (recall that for the normal distribution this is equivalent to setting the covariances to be 0), and variances to be proportional to the corresponding parameter values. 

```python
proportion = 0.5
```

```python
C_M = np.diag(proportion*m_prior)
print('e.g.')
C_M[:5,:5]
```

To find $m_{MAP}$ we solve $$m_{MAP} = \underset{\mathbf{m}}{\text{argmin}}\left|\left| \mathbf{Am} - \mathbf{b} \right|\right|_2^2$$
where $$\mathbf{A} = \begin{bmatrix} 
    C_D^{-1/2}G\\
    C_M^{-1/2}
\end{bmatrix}$$
and
$$\mathbf{b} = \begin{bmatrix} 
    C_D^{-1/2}d\\
    C_M^{-1/2}m_{prior}
\end{bmatrix}$$

Equivalently,
$$m_{MAP} = (\mathbf{A}^T\mathbf{A})^{-1} \mathbf{A}^T \mathbf{b}$$


```python
A = np.concatenate(
    [np.linalg.inv(sqrtm(C_D)) @ G,
     np.linalg.inv(sqrtm(C_M))]
)
b = np.concatenate(
    [np.linalg.inv(sqrtm(C_D)) @ d,
     np.linalg.inv(sqrtm(C_M)) @ m_prior]
)
```

So, the posterior is given by the following (note that $\mathbf{m}_{true}$ and $\mathbf{m}_{prior}$ are also shown for convenience).

```python
m_map = np.linalg.inv(A.T @ A) @ A.T @ b
print("m_true:",  m_true.astype(float))
print("m_prior:", m_prior.astype(float))
print("m_map:",   m_map.astype(float))
```

And the posterior covariance matrix is

```python
C_M_p = np.linalg.inv(G.T @ np.linalg.inv(C_D) @ G + np.linalg.inv(C_M))
C_M_p
```

```python
thing = np.random.multivariate_normal(m_map, C_M_p, size=1000)
prior_thing = np.random.multivariate_normal(m_prior, C_M, size=1000)
```

```python
plt.hist(prior_thing[:,0], bins=40, alpha=0.5);
plt.hist(thing[:,0], bins=40, alpha=0.5, color='k');
```

```python
plt.plot(thing[:,2], thing[:,1], '+');
```

```python
corner(thing);
```

### Reconstructing Data
Here we compare the data produced by the posterior, i.e. $\mathbf{m}_{MAP}$, to that produced by $\mathbf{m}_{true}$. Note that a function of the parameters given by the posterior is called a *posterior predictive distribution* (?).

```python
best_guess = G@m_map
variances = G**2 @ np.diag(C_M_p)
variances = np.diag(G @ C_M_p @ G.T)  # don't know if this is actually correct.......................

plt.figure(figsize=(7,5))
plt.plot(t, G@m_true, label="$\mathbf{Gm}_{true}$", linewidth=2);
plt.errorbar(t, best_guess, yerr=np.sqrt(variances), capsize=3, fmt='+', color="black", markersize=10, label="$\mathbf{Gm}_{MAP}$ (with 1$\sigma$ error bars)");
plt.legend(loc='lower left');
plt.xlabel("Time");
plt.ylabel("Height");
```
