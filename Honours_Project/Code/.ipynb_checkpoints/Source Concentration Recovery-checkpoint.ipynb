# Henry's source reconstruction problem

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as so
```

### Parameters and home-made data

```python
# change these:
m = 100
n = 100
D = 1
v = 1

# x-locations at which we have concentration data (this array should be
# the same length as C)
x = np.linspace(0,300,m)

# times at which we know Cin. Each one of these corresponds to one parameter Cin_i
t = np.linspace(0,250,n)
dt = t[-1] - t[-2]
T = 300

# example Cin
Cin_example = 3 + np.sin(2*np.pi*t/n)*np.sin(0.01*t)
```

* *Why does Oliver's answer use "midpoints?"*

### Constructing the kernal

```python
def forward(x, t):   
    """ Construct kernal (ie. the forward operator)"""
    def k(x,t):    
        """ Kernal function"""
        return x/(2*np.sqrt(np.pi*D*t**3)) * np.exp(-(x-v*t)**2/(4*D*t))
    
    m = len(x)
    n = len(t)
    A = np.zeros([m,n])
    
    for i in range(m):
        for j in range(n):
            A[i,j] = k(x[i], T - t[j])
            
    return A*dt
```

```python
# example forward operator

np.set_printoptions(precision=3)
A = forward(x[:4], t[:4])
A
```

```python
# construct kernal
A = forward(x, t)
```

### Plotting example source data, and the forward problem output

Assume that the following source concentration is the true model.

```python
plt.plot(t,Cin_example);
plt.title("Example source concentration");
plt.xlabel("Time");
plt.ylabel("Concentration");
```

The forward problem output is the spatial distribution of concentration at the fixed time $T$.

```python
# forward output resulting from using Cin_example
output_example = A@Cin_example

# solution of forward problem using example Cin
plt.plot(x,output_example);
plt.title("Spatial distribution at $t=T$\nie. the resulting forward output");
plt.xlabel("x");
plt.ylabel("Concentration");
```

### Add a bit of noise to create our "synthetic" input into the inverse problem
Assume now that the data we collected has a small amount of random noise. Perhaps this came through our inaccurate measurement devices. 

The inverse problem here has the spatial concentration at fixed $T$ as input; whereas, before, the forward problem had the temporal source concentration data as input.

```python
# add a bit of noise to output, to create our own "sythetic" data
output_example_noisy = output_example + np.random.normal(0,0.01,size=m)

plt.plot(x,output_example_noisy);
plt.title("Synthetic data");
plt.xlabel("x");
plt.ylabel("Concentration");
```

### "Recovering" input data (the naive approach)
Notice the size of the values on the y-axis!

```python
# try to recover original source concentration with out synthetic data. We would expect something similar to what we 
# had, though we'll see that the noise added will prove to be a problem. Note that the PSEUDO-inverse is calculated 
# here.
recovered = np.linalg.pinv(A) @ output_example_noisy

plt.plot(t,Cin_example,label="What we expected")
plt.plot(t,recovered,label="'Recovered'");
plt.title("'Recovered' source data");
plt.xlabel("t");
plt.ylabel("Concentration");
plt.legend();
```

```python
thing = A@recovered
plt.plot(x,thing)
plt.plot(x,output_example)
```

### Using a genetic algorithm to find appropriate $C_{in}$

***Note that this may take a few minutes to run.***

Here we are using a metaheuristic to find a good $C_{in}$. We are now essentially running the forward problem many times with "smartly" chosen parameters, to hopefully end up with some decent parameters.

Notice that we *do* get parameter output that results in a seemingly good fit for the spatial concentration distribution; however, see the cell below to see that the source concentration ($C_{in}$), which is what we're really interested in finding, doesn't look right at all (though the scale of the values is roughly correct). 

* *Is the problem we are seeing here "non-uniqueness?"* - Yes, because we see there are more than one possible $C_{in}$'s that produce roughly the same forward problem output.

```python
if 0:

    # data to use
    spacial_data = output_example_noisy

    # least squares
    def fn(x): return ((spacial_data - A@x)**2).sum()

    bounds = [(0,5)]*n
    out = so.differential_evolution(fn,bounds, maxiter=250)
    resulting_Cin = out['x']

    plt.plot(x,spacial_data, label="$C(x,T)$");
    plt.xlabel("x");
    plt.ylabel("Concentration");
    plt.plot(x,A@resulting_Cin, label="Forward model output");
    plt.legend();
    print(out)
```

```python
# initial concentration (as function of time) heuristic output (best guess). Compare this result to the initial
# "Cin_example."
plt.plot(t,resulting_Cin, label="Inverse problem output");
plt.plot(t,Cin_example, label="What we were hoping for");
plt.xlabel('t');
plt.ylabel('Concentration');
plt.title("Source concentration as function of time\n(inverse problem output)");
plt.legend();
```
