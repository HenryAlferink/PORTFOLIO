#### Problem settings

```python
# data standard deviation
data_std = 0.1

# prior standard deviation
prior_std = 1.23

# number of parameters:
n = 100 
# number of initial data points (there will be some additional points added later)
num_initial_points = 3

T = 300 # time at which we want to evaluate the concentration as a function of distance

infeasible = 65 # value below which cost will be infinite
```

#### Creating true parameter set $\mathbf{m}_{true}$

```python
# independent variable corresponding to parameter set:
t = np.linspace(0,250,n)

# distance between time steps:
dt = t[-1] - t[-2] 
```

```python
# Ollie's version...
m_true = 3 + np.sin(2*np.pi*t/n)*np.sin(0.01*t)
```

#### Creating prior parameter set $\mathbf{m}_{prior}$

```python
# constant prior:
m_prior = np.ones(t.shape) * 3
```

```python
# comparing our prior model and the "true" model
plt.figure(figsize=(6,4))
plt.plot(t, m_true, "--", label="True model");
plt.plot(t, m_prior, label="Prior model");
plt.xlabel("Time (t)");
plt.ylabel("Source Concentration ($C_{in}$)");
plt.title("Prior model and true model")
plt.legend();

plt.close()
```

#### Prior covariance matrix

```python
C_prior = prior_std**2 * np.eye(n)
std = np.ones(n) * np.sqrt(prior_std)
Std = np.outer(std, std)
char_L = 10 # characteristic length
def exp_squared(i,j,L): 
    return np.exp(-(i-j)**2/(2*L**2))
C_prior = np.fromfunction(exp_squared, (n,n), L=char_L) * Std + np.eye(n)*1e-4

# plt.imshow(C_prior);
```

#### Forward model

```python
# forward operator parameters
D = 1; v = 1

def kernel(x, t): 
    return x / (2*np.sqrt(np.pi*D*t**3)) * np.exp(-(x-v*t)**2 / (4*D*t))

# construct design matrix. Note that this matrix is a function of T, the time we observe the data vector.
def design(x, t):
    m = x.size; n = len(t)
    G = np.zeros((m, n))
    for i in range(m):
        for j in range(n):
            G[i,j] = kernel(x[i], T-t[j]) * dt
    return G
```

#### Set of measurement locations $\mathcal{X}$

```python
# initialize data class, starting with an assumed linear spacing of already-known data points 
data = h.data(
    x_lo=20, 
    x_hi=300, 
    num_data_points=num_initial_points, 
    design=design, 
    t=t, 
    m_true=m_true, 
    m_prior=m_prior,
    data_std=data_std,
    C_prior=C_prior,
)
```

```python
data.clear_additional()

# New possible measurement locations to add:
newpoints = np.linspace(0,300,100)

# add new possible measurement locations. This also automatically computes the corresponding concentration data
# at that location
for i in range(len(newpoints)):
    data.add_points([newpoints[i]])
```

#### Measurement location cost function

```python
# make sure to give this function the same arguments given here
def cost_function(x, infeasible):
    """ Function that returns a vector of cost values associated with the given input measurement locations."""
    c = np.zeros(x.shape)
    number_infeasible = (x<=infeasible).sum() # number of values in vector below infeasible point
    c[x<=infeasible] = 99999 # infinite cost for x-values below a certain distance
    remaining = len(x) - number_infeasible # remaining number of locations in vector   
    
    curr_pos = number_infeasible
    levels = np.array([3,2,1])*0.55 # different possible cost values
    shift = np.floor(remaining / len(levels)).astype('int')
    for i in range(len(levels)):
        c[curr_pos:] = levels[i]
        curr_pos += shift
    return c

# zero cost function
def cost_function(x, infeasible):
    return np.ones(x.shape)*0
```

```python
plt.plot(data.X_plot, data.d_plot, label="True concentration data");
plt.plot(data.X_plot, cost_function(data.X_plot, infeasible), linewidth=3, label="Cost function");
plt.ylim((0, max(data.d_plot)+(max(data.d_plot)-min(data.d_plot))*0.15));
plt.xlabel('x');
plt.title("True data plotted with cost function on same axis")
plt.legend();
```

```python
# Evalulate costs within data object for each new possible measurement location
data.evaluate_cost(cost_function, infeasible)
```
