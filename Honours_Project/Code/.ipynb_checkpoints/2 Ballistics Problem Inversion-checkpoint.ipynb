# Ballistics Problem Inversion

In this notebook we show two different methods of finding the solution to the inverse problem: the least squares method in the frequentist framework, and the method that uses the Bayesian framework.

If you want to play around, I recommend changing the parameters `data_stddev`, and you could also play around with changing the prior covariance matrix `C_M`.

```python
%matplotlib inline

import numpy as np
from scipy.linalg import sqrtm
import matplotlib.pyplot as plt
import matplotlib
from corner import corner
import IPython

np.random.seed(10)

# plt.rcParams.update({'font.size': 18})
matplotlib.rcdefaults()

# for outputting in .pgf format
matplotlib.rcParams.update({
    'font.family': 'serif',
    'text.usetex': True,
    'pgf.rcfonts': False,
})
```

## Ballistics Example
### Setup
We want to solve $\mathbf{d} = \mathbf{Gm}$ for $\mathbf{m}$ with the regression model being
$$y(t) = m_1 + m_2 t - \frac{1}{2}m_3 t^2.$$
This is the equation used for simple projectile motion.

For example purposes, we will use $\mathbf{m}_{true}$ to be

```python
m_true = np.array([600, 100, 9.8])
```

And we'll set the time steps to be

```python
m = 25
```

```python
t = np.arange(1,m+1)
print('e.g.,')
t[:10]
```

With $m_{true}$ we can create some synthetic data for different time steps. We add a bit of noise to each data point; i.e. we are using
$$d_i = y(t_i) + \epsilon_i$$
where $\epsilon_i \overset{iid}{\sim} N(0,64)$. Note that we make the synthetic data from samples taken from a normal distribution around the *true* model.

```python
data_stddev = 300 # standard deviation of data points
```

```python
def f(time, model): return model[0] + model[1]*time - 0.5*model[2]*time**2 

d = f(t, m_true) + np.random.normal(0, data_stddev, m) # the second argument is std. deviation, not variance.
```

We now construct the $G$ matrix as follows
$$G_{i,.} = [1, t_i, -(1/2)t_i^2]$$

```python
G = np.zeros((m, 3))
for i in range(m):
    G[i,:] = [1, t[i], -0.5*t[i]**2]
print('e.g.')
G[:5, :]
```

### Solving using Least Squares

$m_{ls} = (G^TG)^{-1}G^Td$ solves in the least squares sense.

```python
# least squares solution
m_ls = np.linalg.inv(G.T@G)@G.T@d
```

```python
plt.figure(figsize=(6.50127,3))

# plt.plot(t, f(t, m_true), label="True Model");
# plt.errorbar(x=t, y=d, yerr=data_stddev, fmt=".", capsize=3, label="Synthetic Data (with 1$\sigma$ error bars)");

plt.plot(t,d,'.',label='data',c='k')
plt.plot(t, f(t, m_true), label="true Motion of Projectile",c='k');
plt.plot(t, f(t,m_ls), label="least Squares Estimate",c='tab:orange')
plt.xlabel('Time');
plt.ylabel('Height');
plt.legend(loc="lower left");
plt.tight_layout()
# plt.savefig('Figures/Projectile/1.pgf')
```

### Solving using the Bayesian framework
The above was done in the frequentist framework. We now do the same thing but in the Bayesian framework. Now we take the solution to be a probability distribution rather than a fixed value. We assume that the data is iid normal distributed. This means the data covariance matrix will be
$$C_D = \sigma_d^2 \mathbf{I}_m.$$

```python
# data covariance
C_D = data_stddev**2 * np.eye(m)
print('e.g.')
C_D[:5,:5]
```

We will take the _prior mean_ to be 

```python
np.array([600, 100, 9.8])
m_prior = np.array([550, 110, 5])
```

and we set the prior covariance matrix such that variables are independent (recall that for the normal distribution this is equivalent to setting the covariances to be 0), and we take the standard deviations (square roots of the diagonal entries of the covariance matrix) such that the true parameter values still lie somewhere in the range of one standard deviation from the prior mean. Obviously this is a contrived example here, but in the real world situation it _is_ important that we get the somewhat right (or set the standard deviations well-above our threshold of comfort--though this will mean our prior less and less meaningful).

```python
proportion = 0.5
```

```python
C_M = np.diag([3000, 180, 30])
print('e.g.')
C_M[:5,:5]
```

To find $m_{MAP}$, which, in our case of having a normally distributed prior and normally distributed data noise will be the mean of the (also normally distributed) posterior, we solve $$m_{MAP} = \underset{\mathbf{m}}{\text{argmin}}\left|\left| \mathbf{Am} - \mathbf{b} \right|\right|_2^2$$
where $$\mathbf{A} = \begin{bmatrix} 
    C_D^{-1/2}G\\
    C_M^{-1/2}
\end{bmatrix}$$
and
$$\mathbf{b} = \begin{bmatrix} 
    C_D^{-1/2}d\\
    C_M^{-1/2}m_{prior}
\end{bmatrix}$$

Or equivalently,
$$m_{MAP} = (\mathbf{A}^T\mathbf{A})^{-1} \mathbf{A}^T \mathbf{b}$$
Notice that this takes on the same form as finding the least squares solution in the frequentist framework; this is what we get under the particular assumptions we have taken; however, we are certainly not working in the frequentist framework here!

```python
A = np.concatenate(
    [np.linalg.inv(sqrtm(C_D)) @ G,
     np.linalg.inv(sqrtm(C_M))]
)
b = np.concatenate(
    [np.linalg.inv(sqrtm(C_D)) @ d,
     np.linalg.inv(sqrtm(C_M)) @ m_prior]
)
```

So, the posterior mean is given by the following (note that $\mathbf{m}_{true}$ and $\mathbf{m}_{prior}$ are also shown for convenience).

```python
m_map = np.linalg.inv(A.T @ A) @ A.T @ b
print("m_true:",  m_true.astype(float))
print("m_prior:", m_prior.astype(float))
print("m_map:",   m_map.astype(float))
```

And the posterior covariance matrix is

```python
C_M_p = np.linalg.inv(G.T @ np.linalg.inv(C_D) @ G + np.linalg.inv(C_M))
C_M_p
```

```python
thing = np.random.multivariate_normal(m_map, C_M_p, size=1000)
prior_thing = np.random.multivariate_normal(m_prior, C_M, size=1000)
```

Below we plot the "true parameter". However, note that in the Bayesian view there is no such thing as a "true parameter", but only a "true parameter mean". Notice that the prior is the most uncertain distribution ("solution"), and that the posterior "updates" the prior distribution to be closer to the true mean. This is particularly clear in the third parameter.

```python
for i in range(len(m_true)):
    plt.hist(prior_thing[:,i], bins=40, alpha=0.3, label="prior", color="k");
    plt.hist(thing[:,i], bins=40, alpha=0.6, color='tab:red', label="posterior");
    plt.axvline(m_true[i], c="k", linewidth="3", label="true mean")
    plt.title(f"Parameter $m_{i+1}$")
    plt.legend();
    plt.show()
```

### Reconstructing Data
Here we compare the data produced by the posterior, i.e. $\mathbf{m}_{MAP}$, to that produced by $\mathbf{m}_{true}$. Note that a function of the parameters given by the posterior is called a *posterior predictive distribution*.

```python
plt.figure(figsize=(6.50127,3))

d_map = f(t,m_map) # "pushed forward" posterior mean (pushed forward into data-space that is)
C_pushed = G @ C_M_p @ G.T # "pushed forward" posterior covariance matrix

plt.plot(t,d,'*',label='data',c='k', markersize=2)
plt.plot(t,f(t,m_true),label="true data space mean",c="tab:orange")
plt.plot(t,f(t,m_prior),label="prior data space mean")
plt.plot(t, d_map,label="posterior data space mean",c="k");
plt.plot(t,d_map + 1.96*np.sqrt(np.diag(C_pushed)), ":", c="k")
plt.plot(t, d_map - 1.96*np.sqrt(np.diag(C_pushed)), ":", c="k",label="95\% credible region")
plt.title("Data space solution: compare with least squares version!")
plt.xlabel("time")
plt.ylabel("height")
plt.legend();
```
