# Source Inversion Example
In this problem we know the concentration of a pollutant at multiple distances $\mathcal{X}=\{x_i:i=1,2,\ldots,m\}$ away from the pollutant source. The pollutant source is located at $x=0$. Notice that in this problem the parameters are directly the thing we are wanting to find, whereas in the ballistics problem (other notebook) the thing we wanted to find was a *function* of the parameters and not just the parameters themselves. We are trying to find the history of pollution outflow at $x=0$ as a function of time.

### Parameters (change these!)

```python
# data standard deviation (don't ask why I'm using standard dev here and variance for the prior...).
# note that this notebook is set up so that the true standard deviation of the data is the same as the 
# standard deviation we use for the data covariance matrix, i.e., we use the correct value. In future, it
# might be good to relax this: to show what happens when you make a wrong assumption of standard deviation.
data_std = 0.1

# prior variance. 'constant' just means it's the same for all parameters (which is actually also true for data_std...)
prior_constant_variance = 1

# number of parameters:
n = 100 

# number of initial measurement locations
num_initial_points = 3

# time at which we want to evaluate the concentration as a function of distance
T = 300 

# forward operator parameters (prob not worth messing around with these...)
D = 1
v = 1

# choose which set of additional measurement locations you want to add. Choose either 1, 2, 3, or None.
additional_set = None
```

#### Creating parameter set $\mathbf{m}_{true}$

For example purposes, here we create a "true" parameter set which we will later add noise to, so that we can try to reconstruct the original again.

```python
# independent variable corresponding to parameter set:
t = np.linspace(0,250,n)

# distance between time steps:
dt = t[-1] - t[-2] 
```

```python
# m_true = 10 - 9.7*(0.05*t - 4.9) + 8*(0.05*t - 4.9)**4 + 1.3*(0.05*t - 4.9)**5 - 0.1*(0.05*t - 4.9)**8
# m_true[m_true<0] = 0
# m_true = m_true/m_true.sum()*20

# Ollie's version...
m_true = 3 + np.sin(2*np.pi*t/n)*np.sin(0.01*t)
```

```python
# plt.figure(figsize=2*np.array((6,4)))
plt.figure(figsize=(6.50127,3))
plt.plot(t,m_true);
plt.xlabel("Time (t)");
plt.ylabel("Source Concentration $C_{in}(t)$");
# plt.title("True parameter set (a function of time)");
plt.tight_layout()
# plt.close()
```

#### Forward model

Above we made an assumption of what the true parameter set $\mathbf{m}$ looks like. Now we create the forward operator, that is, $\mathbf{G}$ such that 

$$\mathbf{d}_{true}=\mathbf{Gm}_{true}.$$

Note that finding the operator is not the main focus of this project. We assume it is given.

We now apply the forward model on this true source history to get the "true" concentration data at different distances away from the source (note that "true" here relys on the assumption that the forward operator gives an exact result--which in practice is seldom the case), i.e. we find $\mathbf{d}_{true}$. We then apply a bit of noise in order to mimic collected field data.

```python
def kernel(x, t): 
    return x / (2*np.sqrt(np.pi*D*t**3)) * np.exp(-(x-v*t)**2 / (4*D*t))

# construct design matrix. Note that this matrix is a function of T, the time we observe the data vector.
def design(x, t):
    m = len(x); n = len(t)
    G = np.zeros((m, n))
    for i in range(m):
        for j in range(n):
            G[i,j] = kernel(x[i], T-t[j]) * dt
    return G
```

#### Set of measurement locations $\mathcal{X}$

Here we define the spatial locations at which we choose to collect our concentration data $y_i$. Note that this set of collected concentration data will create the vector $\mathbf{d}$, with $d_i=C(x_i)$, $x_i \in \mathcal{X}$. Later on we'll try to find an optimal measurement location set $\mathcal{X}$ such that parameter uncertainty is minimized. Choosing this optimal set constitutes data-worth analysis.

We start with an initial set of data which we assume was already known before the experiment was done. And we also define a point (or set of points) that is a candidate for being added (see data-worth analysis).

Note: when adding additional points, it's probably not a good idea to add points that are already in the initial set of points...

```python
# initialize data class, starting with an assumed linear spacing of data points 
data = h.data(
    x_lo=80, 
    x_hi=290, 
    num_data_points=num_initial_points, 
    design=design, 
    t=t, 
    m_true=m_true, 
    m_prior=None,
    data_std=data_std,
    C_prior=None,
)

# add some potential datapoints. Note that the order that these methods are called means the given
# list will have that associated index.
# data.add_points(x=[30,40,50,55,70])
# data.add_points(x=[120,250])

if additional_set==1:
    data.add_points(x=[90,95,100,115,120])
    # Choose which datapoint set to (see plot above). SET AS None IF NO DATAPOINTS ARE SELECTED!
    additional_dataset_number = 1 
elif additional_set==2:
    data.add_points(x=[240,250,260,275,295])
    additional_dataset_number = 1
elif additional_set==3:
    data.add_points(x=[90,95,100,115,120,240,250,260,275,295]) # set 2 and 3 together
    additional_dataset_number = 1
else:
    additional_dataset_number = None
```

### Prior
Now, given only the noisy data (and not the true function as shown above), we can use the Bayesian framework to solve. We start by choosing a prior.

```python
# creating the (somewhat arbitrary) "Batman" prior model
# shift = 20
# m_prior = 100 + 0.1*(t-100-shift) - 0.1*(t-100-shift)**2
# where_zero = (m_prior<0) # boolean for where m_prior is currently below 0 (this will be used later)
# m_prior[where_zero] = 0
# m_prior = m_prior/m_prior.sum()*15
# # m_prior[(t>90+shift)&(t<110+shift)] = 0.35 # hump
# m_prior = m_prior + 3

# sine wave prior
# shift = -9.3
# shift = 2
# m_prior = 0.8*np.sin(9.89*t-shift) + 3

# constant prior:
m_prior = np.ones(t.shape) * 3
```

```python

# comparing our prior model and the "true" model
# plt.figure(figsize=2*np.array((6,4)))
plt.figure(figsize=(6.50127,3))

plt.plot(t, m_true, "--", label="True Source Reconstruction");
plt.plot(t, m_prior, label="Prior");
plt.plot(t, m_prior + 1.96*np.sqrt(prior_constant_variance), "k--")
plt.plot(t, m_prior - 1.96*np.sqrt(prior_constant_variance), "k--", label="95\% Credible Interval")
plt.xlabel("Time (t)");
plt.ylabel("Source Concentration $C_{in}(t)$");
# plt.title("Prior model and true model")
plt.legend();
plt.tight_layout()
# plt.close()
```

### Covariance Matrices (data and prior)
Prior covariance and data covariance matrices.

**Question:** Should the data covariance matrix contain the same values along the diagonals corresponding to the variances of each individual data point (data_std in this case)? What should the non-diagonals be?

```python
# covariance of prior. Here we set it as a diagonal matrix so that each component of the 
# prior model has the same variance.
C_p = prior_constant_variance * np.eye(n)

# covariance of data 
C_d = (data_std)**2 * np.eye(data.m)
```

```python
# CONFIG #1: setting the part of the model that is 0 to have a different variance to the rest.
# C_p = 2 * np.eye(n)
# C_p[where_zero] = 0.0001
# for i in range(C_p.shape[0]):
#     for j in range(C_p.shape[0]):
#         if i!=j: C_p[i,j] = 0

# CONFIG #2: still a diagonal matrix, but now the variances decrease linearly with time
# C_p = 0.03 * np.eye(n)
# for i in range(C_p.shape[0]):
#     C_p[i,:] = C_p[i,:] * (C_p.shape[0]-i)
# print(np.diag(C_p))

# CONFIG #3: squared exponential correlation function.
# std is the vector of standard deviations for the prior model. Note that this doesn't 
# necessarily have to be a constant vector (feel free to change--the code allows for it).
std = np.ones(n) * np.sqrt(prior_constant_variance)
Std = np.outer(std, std)
char_L = 10 # characteristic length
def exp_squared(i,j,L): 
    return np.exp(-(i-j)**2/(2*L**2))
C_p = np.fromfunction(exp_squared, (n,n), L=char_L) * Std + np.eye(n)*1e-4




# C_p = prior_constant_variance * np.eye(n) # don't use smoothness prior if you uncomment this.
```

```python
# Prior Covariance matrix
fig,ax = plt.subplots(figsize=(6.50127,3));
ax.imshow(C_p);
fig.tight_layout()
# fig.savefig("Figures/imshow1.png", dpi=400)
# plt.close()
```

### Evaluating Posterior

```python
# # Matrix square root method
# A = np.concatenate(
#     [np.linalg.inv(sqrtm(C_d)) @ G,
#      np.linalg.inv(sqrtm(C_p))]
# )
# b = np.concatenate(
#     [np.linalg.inv(sqrtm(C_d)) @ d,
#      np.linalg.inv(sqrtm(C_p)) @ m_prior]
# )

# Cholesky method
A = np.concatenate(
    [np.linalg.inv(np.linalg.cholesky(C_d)) @ G,
     np.linalg.inv(np.linalg.cholesky(C_p))]
)
b = np.concatenate(
    [np.linalg.inv(np.linalg.cholesky(C_d)) @ d,
     np.linalg.inv(np.linalg.cholesky(C_p)) @ m_prior]
)
```

So, the posterior is given by the following (note that $\mathbf{m}_{true}$ and $\mathbf{m}_{prior}$ are also shown for convenience).

```python
m_map = np.linalg.inv(A.T @ A) @ A.T @ b
```

And the posterior covariance matrix is

```python
C_map = np.linalg.inv(G.T @ np.linalg.inv(C_d) @ G + np.linalg.inv(C_p))
```

```python
# compare the POSTERIOR model (the updated version of the prior after considering the data) to the true model
# plt.figure(figsize=(6,4))
plt.figure(figsize=(6.50127,3))

plt.plot(t, m_true, "--", label="True model");
plt.plot(t, m_map, label="Posterior model (mean)");
plt.xlabel("Time (t)");
plt.ylabel("Source Concentration ($C_{in}$)");
plt.legend(loc="upper left");
plt.tight_layout()

plt.close()
```

### Reconstructing Data (i.e. data space solution)
Here we compare the data produced by the posterior, i.e. $\mathbf{m}_{MAP}$, to that produced by $\mathbf{m}_{true}$. Note that a function of the parameters given by the posterior is called a *posterior predictive distribution* (?). Here we plot 
* the true "data"
* the prior "data"
* the posterior "data"

```python
posterior_pred_distr = G @ m_map
prior_pred_distr     = G @ m_prior

# plt.figure(figsize=(6,4))
plt.figure(figsize=(6.50127,3))

plt.plot(x, d_true, label="$\mathbf{Gm}_{true}$", linewidth=2, c='orange');
plt.plot(x, prior_pred_distr, label="$\mathbf{Gm}_{prior}$", linewidth=2)
plt.plot(x, posterior_pred_distr, label="$\mathbf{Gm}_{MAP}$", linewidth=1, c='k')
plt.legend(loc='upper left');
plt.xlabel("Distance from source (x)");
plt.ylabel("Concentration (C)");
plt.tight_layout()

# plt.close()
```

### Simulation: adding some samples to the plots
We now want to generate many samples to give us a good idea of the "confidence regions" for our inferred function. We can do this in two equivalent (and quite similar) ways:
1. Take many samples from $N(\mathbf{Gm}, \mathbf{C_p})$, where $m_{MAP}$ is the posterior mean and $C_P$ is the posterior covariance; now take these samples and "push them through" the forward operator $\mathbf{G}$ and plot. 

2. Directly sample and plot from $N(\mathbf{Gm}_{MAP}, \mathbf{GC}_p\mathbf{G}^T)$.

### Method 1

```python
num_lines = 500

lines = np.random.multivariate_normal(m_map, C_map, size=num_lines)
pushed_forward = np.zeros((num_lines, G.shape[0]))

for i in range(num_lines):
    pushed_forward[i,:] = G @ lines[i,:]

# plt.figure(figsize=(6,4))
plt.figure(figsize=(6.50127,3))

plt.plot(x, pushed_forward.T, c="green", alpha=0.1, linewidth=0.1);
plt.plot(x, posterior_pred_distr, c="k", linewidth=2, alpha=1, label="Posterior mean (MAP)");
plt.plot(x, d_true, label="True", linewidth=3, c='orange');
plt.legend(loc='upper left');
plt.xlabel("Distance from source (x)");
plt.ylabel("Concentration (C)");
plt.tight_layout()

plt.close()
```

### Method 2

```python
num_lines = 100

lines = np.random.multivariate_normal(G@m_map, G@C_map@G.T, size=num_lines)

# plt.figure(figsize=(6,4))
plt.figure(figsize=(6.50127,3))

plt.plot(x, lines.T, c="green", alpha=0.2, linewidth=0.3);
plt.plot(x, posterior_pred_distr, c="k", linewidth=2, alpha=1, label="Posterior mean (MAP)");
plt.plot(x, d_true, label="True", linewidth=3, c='orange', alpha=0.9);
plt.legend(loc='lower right');
plt.xlabel("Distance from source (x)");
plt.ylabel("Concentration (C)");
plt.tight_layout()

# plt.close()
```
