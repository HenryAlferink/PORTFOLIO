# Source Inversion Example
In this problem we know the concentration of a pollutant at multiple distances $\mathcal{X}=\{x_i:i=1,2,\ldots,m\}$ away from the pollutant source. The pollutant source is located at $x=0$. Notice that in this problem the parameters are directly the thing we are wanting to find, whereas in the ballistics problem (other notebook) the thing we wanted to find was a *function* of the parameters and not just the parameters themselves. 

We are trying to find the history of pollution outflow at $x=0$ as a function of time.

For example purposes, let's assume the following is the true source history:

```python
n = 100 # time resolution of source "reconstruction"
m = 6 # spatial resolution of data

T = 300 # time at which we want to evaluate the concentration as a function of distance
```

#### Creating parameter set $\mathbf{m}_{true}$
For example purposes, here we create a "true" parameter set which we will later add noise to, so that we can try to reconstruct the original again.

```python
t = np.linspace(0,250,n)
```

```python
# m_true = 10 - 9.7*(0.05*t - 4.9) + 8*(0.05*t - 4.9)**4 + 1.3*(0.05*t - 4.9)**5 - 0.1*(0.05*t - 4.9)**8
# m_true[m_true<0] = 0
# m_true = m_true/m_true.sum()*20

# Ollie's version...
m_true = 3 + np.sin(2*np.pi*t/n)*np.sin(0.01*t)
```

```python
plt.plot(t,m_true);
plt.xlabel("Time (t)");
plt.ylabel("Source Concentration ($C_{in}$)");
```

#### Set of measurement locations $\mathcal{X}$
Here we define the spatial locations at which we choose to collect our concentration data $y_i$. Note that this set of collected concentration data will create the vector $\mathbf{y}$, with $y_i=C(x_i)$, $x_i \in \mathcal{X}$. Later on we'll try to find an optimal measurement location set $\mathcal{X}$ such that parameter uncertainty is minimized. Choosing this optimal set constitutes data-worth analysis.

#### Forward model

Above we made an assumption of what the true parameter set $\mathbf{m}$ looks like. Now we create the forward operator, that is, $\mathbf{G}$ such that 

$$\mathbf{y}_{true}=\mathbf{Gm}_{true}.$$

Note that finding the operator is not the main focus of this project. We assume it is given.

We now apply the forward model on this true source history to get the "true" concentration data at different distances away from the source (note that "true" here relys on the assumption that the forward operator gives an exact result--which in practice is seldom the case), i.e. we find $\mathbf{y}_{true}$. We then apply a bit of noise in order to mimic collected field data.

```python
# known parameters
D = 1
v = 1

def kernel(x, t): return x / (2*np.sqrt(np.pi*D*t**3)) * np.exp(-(x-v*t)**2 / (4*D*t))
```

```python
x = np.linspace(0,300,m)

# this one is just so the 'true' data can be plotted nicely
x_plot = np.linspace(0,300,300) 

# adding some data at an arbitray location
x_old = x.copy()

x_new_1 = np.array([20,30,40,50,55,70])
x_new_2 = np.array([140,145,150,160,165,170])
x_new_3 = np.array([240,250,260,275,290,295])

# change this depending on which additional dataset you want to actually use for the posterior
x = np.append(x, x_new_3)

m = len(x)

dt = t[-1] - t[-2] # distance between time steps

# construct design matrix. Note that this matrix is a function of T, the time we observe the data vector.
def design(x, t):
    m = len(x); n = len(t)
    G = np.zeros((m, n))
    for i in range(m):
        for j in range(n):
            G[i,j] = kernel(x[i], T-t[j]) * dt
    return G

G = design(x, t)
```

### Made-up data

```python
# data standard deviation
data_std = 0.1
# data_std = 0.00002
# data_std = 0.00002
# data_std = 1000000
# data_std = 0.05
        
d_true = G @ m_true
d = d_true + np.random.normal(0, data_std, m)
d_plot = design(x_plot, t) @ m_true
```

```python
# for plotting purposes (ie. for Ru...)
d_old = design(x_old, t) @ m_true + np.random.normal(0, data_std, len(x_old))

d_plot_1 = design(x_new_1, t) @ m_true + np.random.normal(0, data_std, len(x_new_1))
d_plot_2 = design(x_new_2, t) @ m_true + np.random.normal(0, data_std, len(x_new_2))
d_plot_3 = design(x_new_3, t) @ m_true + np.random.normal(0, data_std, len(x_new_3))
```

```python
plt.figure(figsize=(6,4))
plt.plot(x_plot, d_plot, label="Underlying model", c='k') # underlying model


# normal way:
# plt.plot(x, d, ".", label="Collected data (with noise)", c="orange", markersize=12)


# for Ru:
plt.plot(x_old, d_old, ".", label="Original datapoints", markersize=12) # original datapoints
plt.plot(x_new_1, d_plot_1, "*", label="Additional datapoints 1", markersize=11)
plt.plot(x_new_2, d_plot_2, "*", label="Additional datapoints 2", markersize=11)
plt.plot(x_new_3, d_plot_3, "*", label="Additional datapoints 3", markersize=11)


plt.xlabel("Distance from source (x)");
plt.ylabel("Concentration (C)");
plt.legend();
```

### Prior
Now, given only the noisy data (and not the true function as shown above), we can use the Bayesian framework to solve. We start by choosing a prior.

```python
# creating the (somewhat arbitrary) "Batman" prior model
m_prior = 100 + 0.1*(t-100) - 0.1*(t-100)**2
where_zero = (m_prior<0) # boolean for where m_prior is currently below 0 (this will be used later)
m_prior[where_zero] = 0
m_prior = m_prior/m_prior.sum()*20
m_prior[(t>90)&(t<110)] = 0.35

# constant prior:
m_prior[:] = 3
```

```python
# comparing our prior model and the "true" model
plt.figure(figsize=(6,4))
plt.plot(t, m_true, "--", label="True model");
plt.plot(t, m_prior, label="Prior model");
plt.xlabel("Time (t)");
plt.ylabel("Source Concentration ($C_{in}$)");
plt.title("Prior model and true model")
plt.legend();
```

### Covariance Matrices (data and prior)
Prior covariance and data covariance matrices.

**Question:** Should the data covariance matrix contain the same values along the diagonals corresponding to the variances of each individual data point (data_std in this case)? What should the non-diagonals be?

```python
prior_constant_variance = 1.5 #0.003

# covariance of prior. Here we set it as a diagonal matrix so that each component of the 
# prior model has the same variance.
C_p = prior_constant_variance * np.eye(n)

# covariance of data (a little bit of le-way added...)
C_d = (data_std)**2 * np.eye(m)
```

```python
# CONFIG #1: setting the part of the model that is 0 to have a different variance to the rest.
# C_p = 2 * np.eye(n)
# C_p[where_zero] = 0.0001
# for i in range(C_p.shape[0]):
#     for j in range(C_p.shape[0]):
#         if i!=j: C_p[i,j] = 0

# CONFIG #2: still a diagonal matrix, but now the variances decrease linearly with time
# C_p = 0.03 * np.eye(n)
# for i in range(C_p.shape[0]):
#     C_p[i,:] = C_p[i,:] * (C_p.shape[0]-i)
# print(np.diag(C_p))

# CONFIG #3: squared exponential correlation function.
# std is the vector of standard deviations for the prior model. Note that this doesn't 
# necessarily have to be a constant vector (feel free to change--the code allows for it).
std = np.ones(n) * np.sqrt(prior_constant_variance)
Std = np.outer(std, std)
char_L = 10 # characteristic length
def exp_squared(i,j,L): return np.exp(-(i-j)**2/(2*L**2))
C_p = np.fromfunction(exp_squared, (n,n), L=char_L) * Std + np.eye(n)*1e-4
# C_p[0,:]
```

```python
# Prior Covariance matrix
plt.imshow(C_p);
```

### Evaluating Posterior

```python
# Matrix square root method
A = np.concatenate(
    [np.linalg.inv(sqrtm(C_d)) @ G,
     np.linalg.inv(sqrtm(C_p))]
)
b = np.concatenate(
    [np.linalg.inv(sqrtm(C_d)) @ d,
     np.linalg.inv(sqrtm(C_p)) @ m_prior]
)

# Cholesky method
A = np.concatenate(
    [np.linalg.inv(np.linalg.cholesky(C_d)) @ G,
     np.linalg.inv(np.linalg.cholesky(C_p))]
)
b = np.concatenate(
    [np.linalg.inv(np.linalg.cholesky(C_d)) @ d,
     np.linalg.inv(np.linalg.cholesky(C_p)) @ m_prior]
)
```

So, the posterior is given by the following (note that $\mathbf{m}_{true}$ and $\mathbf{m}_{prior}$ are also shown for convenience).

```python
m_map = np.linalg.inv(A.T @ A) @ A.T @ b
```

And the posterior covariance matrix is

```python
C_map = np.linalg.inv(G.T @ np.linalg.inv(C_d) @ G + np.linalg.inv(C_p))
```

```python
# compare the POSTERIOR model (the updated version of the prior after considering the data) to the true model
plt.figure(figsize=(6,4))
plt.plot(t, m_true, "--", label="True model");
plt.plot(t, m_map, label="Posterior model (mean)");
plt.xlabel("Time (t)");
plt.ylabel("Source Concentration ($C_{in}$)");
plt.legend(loc="upper left");

# don't plot...
plt.close()
```

```python
samples = 4000
# which parameter to look at:
param_no = 60

# note that the output won't look exactly like a probability distribution; 
# but just make the num bins for each distribution to be the sae.

prior_dist = np.random.multivariate_normal(m_prior, C_p, samples)
post_dist = np.random.multivariate_normal(m_map, C_map, samples)

plt.figure(figsize=np.array((6,4)))
plt.hist(prior_dist[:, param_no], bins=70, density=True, label="Prior");
plt.hist(post_dist[:,param_no], bins=70, alpha=0.8, density=True, label="Posterior");
plt.axvline(m_true[param_no], c="red", label="True value", linewidth=4);
plt.xlabel("Parameter");
ax = plt.gca()
ax.yaxis.set_visible(False)
plt.legend();

plt.close()
```

### Reconstructing Data
Here we compare the data produced by the posterior, i.e. $\mathbf{m}_{MAP}$, to that produced by $\mathbf{m}_{true}$. Note that a function of the parameters given by the posterior is called a *posterior predictive distribution* (?). Here we plot 
* the true "data"
* the prior "data"
* the posterior "data"

(How should I actually be referring to these things?)

```python
posterior_pred_distr = G@m_map
prior_pred_distr     = G@m_prior

plt.figure(figsize=(6,4))

plt.plot(x, d_true, label="$\mathbf{Gm}_{true}$", linewidth=2, c='orange');
plt.plot(x, prior_pred_distr, label="$\mathbf{Gm}_{prior}$", linewidth=2)
plt.plot(x, posterior_pred_distr, label="$\mathbf{Gm}_{MAP}$", linewidth=1, c='k')
plt.legend(loc='upper left');
plt.xlabel("Distance from source (x)");
plt.ylabel("Concentration (C)");

plt.close()
```

### Simulation
We now want to generate many samples to give us a good idea of the "confidence regions" for our inferred function. We can do this in two equivalent (and quite similar) ways:
1. Take many samples from $N(\mathbf{Gm}, \mathbf{C_p})$, where $m_{MAP}$ is the posterior mean and $C_P$ is the posterior covariance; now take these samples and "push them through" the forward operator $\mathbf{G}$ and plot.

### Non-Pushed-Forward Version (i.e. parameters plotted directly)

```python
# plotting the model directly
num_lines = 10

lines = np.random.multivariate_normal(m_map, C_map, size=num_lines)

plt.figure(figsize=(6,4))

plt.plot(t, lines.T, c="green", alpha=0.4, linewidth=0.5);
plt.plot(t, m_map, c="k", linewidth=2, alpha=1, label="Posterior mean (MAP)");
plt.plot(t, m_map+2*np.sqrt(np.diag(C_map)), "--", c="k", linewidth=1, label="95% credible interval")
plt.plot(t, m_map-2*np.sqrt(np.diag(C_map)), "--", c="k", linewidth=1)
plt.plot(t, m_true, label="True model", linewidth=3, c='orange');
plt.legend(loc='upper left');
plt.xlabel("Time (t)");
plt.ylabel("Source concentration ($C_{in}$)");
plt.ylim((0,7.5));

props = dict(boxstyle='round', facecolor="white", alpha=0.15)

# NOTE: I'm just taking the REAL part of the -log(det(C_p))...
textstr = "$tr(C_p)/m = $ {:.2f}\n$-log(det(C_p))=$ {:.0f}".format(np.trace(C_map)/n, 
                                                                   -np.sum(np.log(np.linalg.eigvalsh(C_map))))
# textstr = "$tr(C_p)/m = $ %.2f"%(np.trace(C_map)/m)

# place a text box in upper left in axes coords
plt.text(0.62, 0.99, textstr, transform=ax.transAxes,
        verticalalignment='top', bbox=props);
```
